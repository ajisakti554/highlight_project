**Academic Performance KPI**

Ini adalah KPI untuk melihat keberhasilan akademik secara langsung.

1. Average Exam Score (Rata-rata Skor Ujian): Mengukur performa rata-rata seluruh siswa.

2. Passing Rate (%): Persentase siswa yang mencapai skor di atas ambang batas tertentu (misalnya >70).

3. Improvement Index: Selisih antara Previous_Scores dengan Exam_Score. Ini penting untuk melihat apakah siswa mengalami progres atau penurunan.

**KPI 1**

# 1. Data Cleaning & Pre-processing

```python
import pandas as pd
```

```python
import pandas as pd
import numpy as np

# 1. Menangani missing values pada kolom kategorikal
cols_with_nan = ['Teacher_Quality', 'Parental_Education_Level', 'Distance_from_Home']
for col in cols_with_nan:
    df[col] = df[col].fillna(df[col].mode()[0])

# 2. Cek apakah ada skor yang di luar nalar (misal skor 0-100)
print(df[['Previous_Scores', 'Exam_Score']].describe())

# 3. Menghapus baris yang mungkin memiliki skor null (jika ada)
df.dropna(subset=['Exam_Score', 'Previous_Scores'], inplace=True)
```

Nilai exam yang >100 harus dihapus, setelah dicek, hanya ada data yang >100, yaitu 101 (delete solusinya)

```python
df.drop(df[df['Exam_Score'] > 100].index, inplace=True)
```

```python
df['Exam_Score'].value_counts()
```

```python
df.duplicated().sum()
```

# 2.Feature Engineering (Membuat Rumus KPI)

```python

# 2. Membuat kolom 'Is_Pass' (Boolean: True jika >= 70)
# Ini untuk menghitung Passing Rate nanti
df['Is_Pass'] = df['Exam_Score'] >= 70

# 3. Membuat kolom 'Improvement_Score'
df['Improvement_Score'] = df['Exam_Score'] - df['Previous_Scores']

# 4. Membuat kategori performa untuk visualisasi nanti
def categorize_performance(score):
    if score >= 85: return 'Excellent'
    elif score >= 70: return 'Good'
    elif score >= 55: return 'Average'
    else: return 'Below Average'

df['Performance_Category'] = df['Exam_Score'].apply(categorize_performance)

# Cek 5 data teratas
print(df[['Previous_Scores', 'Exam_Score', 'Improvement_Score', 'Is_Pass', 'Performance_Category']].head())
```

# 3. EDA

```python
# 1. Menghitung Metrik Utama
overall_avg = df['Exam_Score'].mean()
passing_rate = df['Is_Pass'].mean() * 100
avg_improvement = df['Improvement_Score'].mean()

# 2. Menghitung Distribusi Kategori (untuk melihat sebaran kemampuan)
performance_dist = df['Performance_Category'].value_counts(normalize=True) * 100

# 3. Menampilkan Laporan KPI
print("========== ACADEMIC PERFORMANCE REPORT ==========")
print(f"Rata-rata Skor Ujian (Current) : {overall_avg:.2f}")
print(f"Rata-rata Skor Sebelumnya      : {df['Previous_Scores'].mean():.2f}")
print(f"Indeks Kemajuan (Improvement)  : {avg_improvement:.2f}")
print(f"Tingkat Kelulusan (Pass Rate)  : {passing_rate:.2f}%")
print("==================================================")
print("\nDistribusi Performa Siswa (%):")
print(performance_dist.round(2))
```

1. Indikator Krisis: Penurunan Performa Kolektif
* Indeks Kemajuan (Improvement) -7.84: Ini adalah temuan paling kritikal. Secara rata-rata, nilai siswa turun hampir 8 poin dibandingkan ujian sebelumnya. Dalam skala pendidikan, penurunan rata-rata sebesar ini di 6.000+ siswa menunjukkan adanya masalah sistemik, bukan sekadar masalah individu.

* Gap Skor: Rata-rata turun dari 75.07 (masuk kategori Good) menjadi 67.23 (turun ke kategori Average).

2. Tingkat Kelulusan (Pass Rate) yang Mengkhawatirkan
* Hanya 24.58% siswa yang lulus: Dengan standar kelulusan (passing grade) 70 yang kita tetapkan, hanya 1 dari 4 siswa yang berhasil lulus.

* Ini adalah informasi kunci bagi pemangku kepentingan (stakeholder). Jika ini adalah sekolah nyata, kepala sekolah harus segera melakukan evaluasi kurikulum atau metode pengajaran.

3. Distribusi Performa: Penumpukan di "Average"
* 75.42% Siswa berada di kategori Average: Mayoritas besar siswa terjebak di level menengah bawah.

* Excellent hanya 0.51%: Hampir tidak ada siswa yang mencapai skor istimewa (>= 85). Ini menunjukkan bahwa sistem pendidikan saat ini mungkin hanya mampu membuat siswa "bertahan" tapi tidak mampu mendorong mereka untuk "unggul".

4. Analisis Statistik (Standard Deviation)
* Jika kamu perhatikan di gambar describe(), Std (Standard Deviation) untuk Exam_Score (3.89) jauh lebih kecil daripada Previous_Scores (14.39).

* Artinya: Dulu nilai siswa sangat bervariasi (ada yang sangat pintar, ada yang sangat kurang). Sekarang, nilai semua siswa menjadi seragam tapi rendah. Nilai mereka "mengumpul" di angka 67.

```python
# Segmentasi Pass Rate berdasarkan Akses Internet
internet_analysis = df.groupby('Internet_Access').agg(
    Avg_Score=('Exam_Score', 'mean'),
    Pass_Rate=('Is_Pass', lambda x: x.mean() * 100)
).round(2)

# Segmentasi Pass Rate berdasarkan Pendapatan Keluarga
income_analysis = df.groupby('Family_Income').agg(
    Avg_Score=('Exam_Score', 'mean'),
    Pass_Rate=('Is_Pass', lambda x: x.mean() * 100)
).round(2)

print("--- Analisis Berdasarkan Akses Internet ---")
print(internet_analysis)
print("\n--- Analisis Berdasarkan Pendapatan Keluarga ---")
print(income_analysis)
```

1. Analisis Deskriptif: Fenomena "Penurunan Massal"
Berdasarkan statistik deskriptif di gambar pertama:

* Standar Deviasi (std) Menyusut: Pada Previous_Scores, std adalah 14.39, namun pada Exam_Score turun drastis menjadi 3.89.

* Artinya: Dulu nilai siswa sangat bervariasi (ada yang sangat pintar, ada yang sangat kurang). Sekarang, nilai semua siswa menjadi sangat seragam (homogen) tapi di level yang rendah. Mereka "berkumpul" di sekitar angka 67.

2. Segmentasi: Mencari "Akar Masalah" (Deep Dive)
Inilah bagian paling krusial yang kamu hasilkan di gambar terakhir. Kita mulai melihat adanya ketimpangan:

* A. Dampak Akses Internet
Siswa dengan internet memiliki Pass Rate 25.16%, sementara yang tidak punya hanya 17.47%.

* Insight: Ada selisih sekitar ~8% pada tingkat kelulusan. Ini menunjukkan bahwa infrastruktur digital merupakan faktor pendukung performa, namun bukan satu-satunya penentu karena rata-rata skor mereka (Avg_Score) tidak berbeda jauh (67.29 vs 66.47).

B. Faktor Ekonomi (Family Income)
* Ada korelasi linier yang jelas di sini:

* High Income: Pass Rate 29.34%

* Medium Income: Pass Rate 25.77%

* Low Income: Pass Rate 21.15%

* Insight: Siswa dari keluarga mampu memiliki peluang lulus ~40% lebih tinggi dibandingkan siswa dari keluarga kurang mampu. Sebagai analis, kita bisa menyimpulkan adanya kesenjangan ekonomi yang memengaruhi performa akademik.
